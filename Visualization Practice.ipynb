```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Matplotlib & Seaborn
# matplotlib is plotting library of python
# provides static graphs/charts 
# each pyplot function makes some change to the graph
# seaborn is built on top of matplotlib and supports plotting charts with minimal code compared to matplotlib
# matplotlib can't be used for 3d adn more complicated graphs so we use seaborn


# Plot can be saved using below command;
# plot.savefig("output.png")
```

```python
# Let's learn Matplotlib and Seaborn through asnwering the following questions.
# 1 (a). How many records belongs to each Pclass. (bar-chart/histogram)
# 1 (b). How many passengers survived. (bar-plot/countplot)
# 1 (c). Percentages of passengers who survived.
# 2. Display statistical summary of Age and Fare. (boxplot)
# 3. Display density distribution of Age. (Swarmplot/Violinplot)
# 3. What is the relation between Age and Fare (Linear/Non-linear/no relationship)? (scatterplot/Pairplot)
# 4. Distribution of Fare. (check if Fare represents bell-shaped curve) (histogram/distplot)
# 5. Identify strongly correlated columns. (Heatmap)
# 6. Display Q1 and Q2 on same plot area (subplots)
```

```python
##### Barplot
# barplot is used to compare categorical measures

# 1 (a). How many records belongs to each Pclass. (bar-chart)

#plt.bar(x = ['3','2','1'], height =[ df[df['Pclass']==3].shape[0], df[df['Pclass']==2].shape[0], df[df['Pclass']==1].shape[0] ])

plt.bar( x = [str(x) for x in df['Pclass'].unique()], height = df['Pclass'].value_counts(), color=['r','g','b'] )
plt.xlabel('PClass')
plt.ylabel('Count')
plt.title('Bar plot of PClass', fontsize=20)
```

```python
sns.countplot(df['Pclass'])
plt.show()
```

```python
# 1 (b). How many passengers survived. (bar-plot/countplot)

sns.countplot(df['Survived'])

```

```python
# 2. Display statistical summary of Age and Fare. (boxplot)
df.describe()

```

```python
plt.figure(figsize=(25,5))
plt.boxplot(df['Age'], vert=False)
plt.xlabel('Age')
plt.show()
```

```python
plt.boxplot(df['Fare'], vert=False)
plt.xlabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(25,5))
sns.boxplot(df['Age'])
plt.xlabel('Fare')
```

```python
# 3. Display density distribution of Age. (Swarmplot/Violinplot)
plt.figure(figsize=(25,5))
#swarm plot is not present in matplotlib
f = sns.swarmplot(df['Age'])
f.set_title('AGE swarmplot', {'fontsize':20} )
#plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(25,5))
#Violin plot is not present in matplotlib
f = sns.violinplot(df['Age'], orient = 'h')
f.set_title('AGE violinplot', {'fontsize':20} )
plt.xticks(np.arange(0,81,5))
plt.show()
```

```python
# 3. What is the relation between Age and Fare (Linear/Non-linear/no relationship)? (scatterplot/Pairplot)

plt.figure(figsize=(10,5))
plt.scatter(x = df['Age'], y = df['Fare'])
plt.title('Scatter plot b/w Age and Fare', fontsize=20)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.scatterplot('Age', 'Fare', hue='Survived', data=df)
plt.show()
```

```python
# 4. Distribution of Fare. (check if Fare represents bell-shaped curve) (histogram/distplot)

plt.hist(df['Fare'], bins=100)
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.distplot(df['Fare'])
```

```python
sns.distplot(df['Age'])
```

```python
df.corr()
```

```python
# 5. Identify strongly correlated columns. (Heatmap)
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(), annot=True)
plt.show()
```
