```python
### Pandas merge

# T1: customer_details => {customer_id,Name, Address, Age, contact, ....}
# T2: transaction_details => {transaction_id, transaction_date, transaction_amount, customer_id_ref}
# joins are of two types : inner join & outer join (left, right and full)

# inner join -> common records between tables (intersection)
# left outer join -> common records + remaining records of the left table
# right outer join -> common records + remaining records of the right table
# full outer join -> common records + remaining records of the both tables (union)


# -> what type of join
# -> common columns
# -> which is your left/right table/dataframe

left_df = pd.DataFrame({ 'id':[1,2,3,4,5],
                        'Name_left': ['Alex', 'Amy', 'Allen', 'Alice','Ayoung'],
                        'subject_id_left':['sub1','sub2','sub4','sub6','sub5']}) 
right_df = pd.DataFrame( {'id':[1,2,3,4,5],
                         'Name_right': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
                         'subject_id_right':['sub2','sub4','sub3','sub6','sub5']})


 
### Adding a new row in left_df
left_df  = left_df.append({'id': 6,'Name_left':'XYZ', "subject_id_left":'sub11'}, ignore_index = True)
right_df  = right_df.append({'id': 11,'Name_right':'ABC', "subject_id_right":'sub7'}, ignore_index = True)
  
```

```python
# 1.Merge on id
# 2.Merge on id & subject_id
# 3.Perform left join on key = id 
# 4.Perform right join on key = id
# 5.Perform outer join on key = id
# 6.Perform left join on keys = (id, subject_id)
# 7.Perform right join on keys = (id, subject_id)
# 8.Perform outer join on keys = (id, subject_id)
```

```python
right_df
```

```python
left_df
```

```python
#1.Merge on id

pd.merge(left_df,right_df, on = 'id', how = 'inner' )
```

```python
left_df.merge(right_df, on='id')
```

```python
#2.Merge on id & subject_id

left_df.merge(right_df, left_on=['id','subject_id_left'], right_on=['id','subject_id_right'])
```

```python
#3.perform left join on key = id

pd.merge(left_df, right_df, on='id', how='left' )
```

```python
# 4.Perform right join on key = id

pd.merge(left_df, right_df, on='id', how='right')
```

```python
# 5.Perform outer join on key = id

pd.merge(left_df, right_df, on='id', how='outer')
```

```python
# 6.Perform left join on keys = (id, subject_id)
#left_df.merge(right_df, left_on=['id','subject_id_left'], right_on=['id','subject_id_right'], how = 'left' )

pd.merge(left_df, right_df, left_on = ['id','subject_id_left'], right_on = ['id','subject_id_right'], how = 'left')
```

```python
# 7.Perform right join on keys = (id, subject_id)

pd.merge(left_df, right_df, left_on = ['id','subject_id_left'], right_on = ['id','subject_id_right'], how = 'right')
```

```python
# 8.Perform outer join on keys = (id, subject_id)
pd.merge(left_df, right_df, left_on = ['id','subject_id_left'], right_on = ['id','subject_id_right'], how = 'outer')
```
